```python
colonnes_numeriques = df.select_dtypes(include='number')

outliers_dict = {}

for col in colonnes_numeriques.columns:
    Q1 = colonnes_numeriques[col].quantile(0.25)
    Q3 = colonnes_numeriques[col].quantile(0.75)
    IQR = Q3 - Q1
    
    outliers = colonnes_numeriques[
        (colonnes_numeriques[col] < Q1 - 1.5 * IQR) |
        (colonnes_numeriques[col] > Q3 + 1.5 * IQR)
    ]
    
    outliers_dict[col] = outliers.shape[0]

outliers_dict
```

```python
df=df.copy()
df['dt']=pd.to_datetime(df['dt'],errors='coerce')
df['dt'].head()
```

```python
df['Year']=df['dt'].dt.year
```

```python
df['Month']=df['dt'].dt.month
```

```python
x=df.groupby('Year')['AverageTemperature'].mean()
print('Global Temperature per Year :',x)
```

```python
year_hot=x.idxmax()
print('Year hot :',year_hot)
year_cold=x.idxmin()
print('Year Cold :',year_cold)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(x, color='red')
plt.title("Évolution globale  temperature Average (1750-2013)")
plt.xlabel("Year")
plt.ylabel("Temperature average (°C)")
plt.grid(True)
plt.show()
```

```python
y=df.groupby('Month')['AverageTemperature'].mean()
print('Global Temperature per Month :',y)
```

```python
month_hot=y.idxmax()
print('Month hot :',month_hot)
month_cold=y.idxmin()
print('Month Cold :',month_cold)
```

```python
plt.figure(figsize=(12,6))
plt.plot(y, color='b')
plt.title("Évolution global  temperature average (1750-2013)")
plt.xlabel("Month")
plt.ylabel("Temperature average (°C)")
plt.grid(True)
plt.show()
```

```python
x_start=x.iloc[0]
x_end=x.iloc[-1]

if x_start>x_end :
    print(' Temperature is warming')
else:
    print(' Temperature is cooling')

print('Start temperature:',x_start,'.2f°c','\nEnd temperature:',x_end,'.2f °c')

```

```python
avr_per_city=df.groupby('City')['AverageTemperature'].mean()
print('Average city per temperature is ',avr_per_city)
max_per_city=avr_per_city.max()
print('Maximum City is :',max_per_city)
```

```python
plt.figure(figsize=(12,6))
plt.plot(avr_per_city, color='b')
plt.title("Évolution global  temperature average (1750-2013)")
plt.xlabel("")
plt.ylabel("Temperature average (°C)")
plt.grid(True)
plt.show()
```

```python
avg_city=df.groupby('City')['AverageTemperature'].mean()
cold_city=avg_city.min()
print('The coldest city is :',cold_city)
```

```python
extreme_hot=df[df['AverageTemperature']>30]
extreme_cold=df[df['AverageTemperature']<-20]

print(f"Many very hot regions : {extreme_hot['City'].nunique()}")
print(f"Many very cold regions : {extreme_cold['City'].nunique()}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(extreme_hot['Longitude'], extreme_hot['Latitude'], color='red', label='Very hot ', alpha=0.5)
plt.scatter(extreme_cold['Longitude'], extreme_cold['Latitude'], color='blue', label='very cold', alpha=0.5)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Extreme clima ")
plt.legend()
plt.show()
```

```python
extreme_counts = pd.concat([extreme_hot, extreme_cold])
events_per_city = extreme_counts.groupby('City').size().sort_values(ascending=False)
print(events_per_city.head(10))  
```

these regions more exposed than others 

```python
extreme_counts_by_year = extreme_counts.groupby('Year').size()

plt.figure(figsize=(12,5))
plt.plot(extreme_counts_by_year, color='purple')
plt.title("Évolution des événements extrêmes au fil des années")
plt.xlabel("Année")
plt.ylabel("Nombre d'événements extrêmes")
plt.grid(True)
plt.show()
```
