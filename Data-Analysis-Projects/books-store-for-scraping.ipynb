**Install the library**

```python
import requests
from bs4 import BeautifulSoup

url="https://books.toscrape.com/"
data=requests.get(url).text
print(data)

```

Extract the HTML code from the web 

```python
soup=BeautifulSoup(data,"html.parser")
```

```python
soup.find("article",class_="product_pod")
```

```python
soup.find("p",class_="instock availability")
```

```python
soup.find_all("title")
```

find all the title in th HTML code

```python
soup.find_all("p")
```

**Select all the paragraph**

```python
soup.find("p").find_all('i')
```

**find the first paragraph with the tag="i"** 


```python
for t in soup.find("p",class_='price_color'):
    print(t.text)
```

```python
soup.find("p","star-rating Two")
```

**find the first paragrapah who have "star-rating Two"**

```python
for i in soup.find('div').find_all("a"):
    print(i)
```
